Removing Outliers

```python
pd.Series(scipy.stats.zscore(aps['Departure Delay'], axis=0))
```

```python
threshold = 0.05
```

```python
scipy.stats.norm.ppf(q=0.95, scale=1, loc=1)
```

```python
def remove_outliers(aps, columns,threshold):
    aps = aps.copy()
    
    #Calculate the lower and upper bound on the z distribution given the threshold value
    lower_bound = scipy.stats.norm.ppf(q=(threshold / 2), loc=0, scale= 1)
    upper_bound = scipy.stats.norm.ppf(q=1 - (threshold / 2), loc=0, scale= 1)
    
    #Calculate z-score of every example in the columns specified
    outlier_df = aps.loc[:, columns].copy()
    zscores = pd.DataFrame(scipy.stats.zscore(outlier_df, axis = 0), index = outlier_df.index, columns = outlier_df.columns)
    
    #get boolean arrays denoting the outliers examples
    lower_outliers = (zscores < lower_bound).any(axis=1)
    upper_outliers = (zscores >= upper_bound).any(axis=1)
    
    #get indicies of all outlier example
    outliers = pd.concat([lower_outliers, upper_outliers], axis=1).any(axis=1)
    return outliers
```

```python
remove_outliers(
    aps = aps, columns = ['Flight Distance', 'Departure Delay', 'Arrival Delay', 'Check-in Service'], threshold=0.05).any(axis=1)
```
